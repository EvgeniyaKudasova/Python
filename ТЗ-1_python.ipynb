Вам предстоит поработать со списком возрастов посетителей фитнес-клубов age_lst

**Задача 1.** Рассчитайте средний возраст по всему списку age_lst

**Задача 2.** Рассчитайте средний возраст для посетителей, которые моложе 50 лет

**Задача 3.** Оберните предыдущий код в функцию, которая на вход принимает список и пороговое значение

**Задача 4.** Рассчитайте среднее значение для всех элементов списка, за исключением тех, которые лежат вне интервала [Q1-1,5*IQR; Q3+1,5*IQR]

**Задача 5 (со звёздочкой).** Оберните предыдущий код в функцию, которая принимает на вход список и рассчитывает средний возраст, исключая из расчётов выбросы

```python
age_lst = [23, 45, 18, 80, 22, 25, 34, 37, 41, 88, 19, 20, 18, 45, 41, 39, 31, 32, 22, 28, 37]
```

```python
import numpy as np
```

```python
avg_age = sum (age_lst)/ len (age_lst)
print ("Средний возраст посетителей", round (avg_age, 1), 'лет')
```

```python
sum_age = 0
cnt_age = 0
for i in age_lst:
    if i < 50:
        sum_age += i
        cnt_age += 1
avg_age = sum_age/ cnt_age
print ("Средний возраст посетителей, которые моложе 50 лет -", round (avg_age, 1))
```

```python
def avg_age (lst, thrash):
    sum_age = 0
    cnt_age = 0
    for i in lst:
        if i < thrash:
            sum_age += i
            cnt_age += 1
    avg_age = sum_age/ cnt_age
    print ("Средний возраст посетителей, которые моложе", thrash, 'лет', round  (avg_age, 1))
```

```python
avg_age (age_lst, 40)
```

```python
q1 = np.percentile (age_lst, 25)
q3 = np.percentile (age_lst, 75)
iqr = q3 - q1
lower_bound = q1 - 1.5*iqr
upper_bound = q3 + 1.5*iqr
sum_age = 0
cnt_age = 0
for i in age_lst:
    if (i < upper_bound) and (i > lower_bound):
        sum_age += i
        cnt_age += 1
avg_age = sum_age/ cnt_age
print ("Новый средний возраст посетителей", round  (avg_age, 1))
```

```python
def calc_avg_age_adv (lst):
    q1 = np.percentile (lst, 25)
    q3 = np.percentile (lst, 75)
    iqr = q3 - q1
    lower_bound = q1 - 1.5*iqr
    upper_bound = q3 + 1.5*iqr
    sum_age = 0
    cnt_age = 0
    for i in lst:
        if (i < upper_bound) and (i > lower_bound):
            sum_age += i
            cnt_age += 1
    avg_age = sum_age/ cnt_age
    print ("Средний возраст посетителей -", round (avg_age, 1))
    return round (avg_age, 1)
```

```python
calc_avg_age_adv (age_lst)
```
